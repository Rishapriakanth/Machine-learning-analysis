```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram,linkage
from matplotlib import pyplot as plt
X = np.array([[1,4],[2,8],[5,10],[12,18],[14,28]])
# Single Linkage method
Z = linkage(X,'single')
fig = plt.figure(figsize=(5,5))
dn = dendrogram(Z)
plt.xlabel('Samples')
plt.ylabel('Height')
plt.title('Single Linkage Method')
# Complete Linkage method
Z = linkage(X,'complete')
fig = plt.figure(figsize=(5,5))
dn = dendrogram(Z)
plt.xlabel('Samples')
plt.ylabel('Height')
plt.title('Complete Linkage Method')
# Average Linkage method
Z = linkage(X,'single')
fig = plt.figure(figsize=(5,5))
dn = dendrogram(Z)
plt.xlabel('Samples')
plt.ylabel('Height')
plt.title('Average Linkage Method')
```

```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
data = np.array([[8,12],[12,17],[20,20]])
groups = AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='single')
print("The Predicted groupings for test data")
print("Test data are (8,12),(12,17),(20,20)")
pred = groups.fit_predict(data)
print("\n Cluster ID's are")
print(pred)
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram,linkage
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
iris = pd.read_csv('/content/Iris.csv')
iris.drop(['Species'],axis=1,inplace=True)
iris.drop(['SepalLengthCm'],axis=1,inplace=True)
iris.drop(['SepalWidthCm'],axis=1,inplace=True)
gps= AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
gps.fit_predict(iris)
plt.scatter(iris['PetalLengthCm'],iris['PetalWidthCm'],c=gps.labels_,cmap='cool')
plt.xlabel('Samples')
plt.ylabel('Height')
plt.title('Complete Linkage Method')
```
